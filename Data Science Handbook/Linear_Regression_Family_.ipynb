## Theoretical Framework

### What Is Regression Analysis?

Regression analysis is a statistical technique used to study the relationship between one dependent variable and one or more independent variables. It allows us to understand how changes in predictor variables are associated with changes in the outcome variable.

Regression models are widely used for two main purposes:

- Explanation: understanding how different factors influence an outcome.
- Prediction: estimating future or unknown values based on observed data.

For example, economists may use regression analysis to examine how education, unemployment, and inflation are related to human development outcomes across countries.

### Why Is Regression Important?

Regression is one of the most widely used methods in data science because it offers several advantages:

- It provides interpretable results.
- It is computationally efficient.
- It serves as the foundation for many advanced machine learning methods.
- It allows researchers to quantify relationships between variables.
- It supports evidence-based decision-making.

Despite the availability of more complex algorithms, regression models remain essential tools in economics, business, healthcare, and scientific research.

Simple Linear Regression examines the relationship between one predictor variable and one target variable.

####2. Multiple Linear Regression

Multiple Linear Regression extends simple regression by incorporating several predictor variables simultaneously.

### Residuals and Prediction Errors

A residual is the difference between an observed value and the value predicted by the model.

Residuals help us evaluate how well a regression model fits the data.

Small residuals indicate that predictions are close to the observed values, whereas large residuals suggest that the model may not adequately capture the underlying relationships.

Analyzing residuals is an important part of regression diagnostics because it can reveal potential issues such as nonlinearity, heteroscedasticity, and influential observations.

### Assumptions of Linear Regression

#### Linearity

The relationship between the predictors and the target variable should be approximately linear.

This means that changes in the predictors are associated with proportional changes in the outcome variable. If the relationship is strongly nonlinear, a simple linear model may fail to capture the underlying pattern.

*Figure 3.1. Comparison between approximately linear and nonlinear relationships.*

#### Independence of Observations

Observations should be independent of one another. In other words, the value of one observation should not directly influence another observation.

Violations of this assumption are common in time series data, where observations are often correlated over time.

*Figure 3.2. Independent observations compared with observations exhibiting dependence over time.*

#### Homoscedasticity

The variance of residuals should remain relatively constant across all predicted values.

When the spread of residuals changes substantially, heteroscedasticity is present.

*Figure 3.3. Constant residual variance compared with increasing residual variance.*

#### Normality of Residuals

Residuals are assumed to follow an approximately normal distribution.

This assumption is particularly important when conducting statistical inference.

image.png

*Figure 3.4. Residuals following an approximately normal distribution.*

#### No Perfect Multicollinearity

Predictor variables should not be perfectly correlated with one another.

Highly correlated predictors can make coefficient estimates unstable and difficult to interpret.

image.png

*Figure 3.7. Comparison between weakly correlated and highly correlated predictors.*

Multicollinearity is commonly assessed using the Variance Inflation Factor (VIF), defined as:

$$
VIF_j = \frac{1}{1 - R_j^2}
$$

where $(R_j^2)$ is obtained by regressing predictor $(j)$ on the remaining predictors.

As a general guideline:

- VIF ≈ 1 indicates little or no multicollinearity.
- VIF between 1 and 5 suggests moderate correlation.
- VIF above 5 may indicate problematic multicollinearity.
- VIF above 10 is often considered severe multicollinearity.

When multicollinearity is substantial, possible solutions include removing redundant variables or applying regularization methods such as Ridge Regression and Elastic Net Regression.

## Mathematical Foundations

## The Best Fit Line

When a scatter plot contains many observations, several different lines could be drawn through the data. However, some lines describe the relationship better than others.

The objective of linear regression is to identify the line that best represents the overall pattern of the data. This line is known as the **best fit line**.

The best fit line minimizes the differences between the observed values and the values predicted by the model. These differences are called residuals.

Rather than passing through every observation, the best fit line provides the most balanced representation of the relationship between the predictor and the outcome variable.

image.png

*Figure 3.8. The best fit line summarizes the overall relationship between the predictor and the outcome variable.*

## Simple Linear Regression Equation

Once the best fit line has been identified, it can be described mathematically using the simple linear regression equation:

##$$
y = β₀ + β1x + ε
$$

where:

- y  is the dependent variable (target),
- x  is the independent variable (predictor),
- β₀ is the intercept,
- β1 is the slope coefficient,
- ε  is the error or the residuals

The intercept represents the expected value of the target variable when the predictor equals zero, whereas the slope represents the expected change in the target variable associated with a one-unit increase in the predictor.

####Key Idea
The best fit line is the foundation of linear regression.

First, we identify the line that best represents the data. Then, we describe this line mathematically using the regression equation. In the next section, we will learn how the model determines which line is the best by minimizing prediction errors through the Least Squares method.

## Least Squares Estimation

If many different lines can be drawn through the data, how does linear regression determine which line is the best?

Linear regression uses the **Least Squares Method** to answer this question.

The idea is simple: the best fit line is the one that minimizes the differences between the observed values and the values predicted by the model.

These differences are called **residuals**.

For each observation, the residual is defined as:

##$$
e_i = y_i - \hat{y}_i
$$

where:

- $e_i$ is the residual for observation $i$,
- $y_i$ is the observed (acual) value,
- $\hat{y}_i$ is the predicted value.

Because some residuals are positive and others are negative, simply summing them would cause them to cancel each other out.

To avoid this problem, residuals are squared before being added together.

The line that produces the smallest total squared error is selected as the best fit line.

###Least Squares Estimation

image.png

*Figure 3.10. Residuals represent the vertical distances between observed values and predictions. The least squares method selects the line that minimizes the total squared residuals.*

### Residual Sum of Squares (RSS)

The total prediction error of a regression model is commonly measured using the Residual Sum of Squares (RSS).

RSS is defined as:

##$$
RSS = \sum_{i=1}^{n}\left(y_i-\hat{y}_i\right)^2
$$

where:

- $n$ is the number of observations,
- $y_i$ is the observed value,
- $\hat{y}_i$ is the predicted value.

A smaller RSS indicates that the predictions are closer to the observed values and that the fitted line better represents the data.

### Mean Squared Error (MSE)

Although RSS is useful, its magnitude naturally increases as the number of observations grows.

For this reason, the Mean Squared Error (MSE) is often used to evaluate regression models.

MSE is defined as:

$$
MSE = \frac{1}{n}\sum_{i=1}^{n}\left(y_i-\hat{y}_i\right)^2
$$

A lower MSE indicates better predictive performance because, on average, the model produces smaller prediction errors.

###Why Squaring the Errors

image.png

*Figure 3.11. Squaring residuals prevents positive and negative errors from canceling each other and gives greater emphasis to larger errors.*

####Key Idea
The least squares method transforms the problem of finding the best fit line into an optimization problem.

By minimizing the total squared residuals, linear regression identifies the line that best represents the observed data while reducing prediction errors.

## Multiple Linear Regression

In many real-world problems, a single predictor is often insufficient to explain variations in the target variable.

Multiple Linear Regression extends simple linear regression by incorporating several predictor variables simultaneously.

Instead of relying on only one predictor, the model estimates the combined effect of multiple factors on the outcome variable.

The multiple linear regression model is defined as:

$$
y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \cdots + \beta_px_p + \varepsilon
$$

where:

- $y$ is the dependent variable,
- $x_1, x_2, \ldots, x_p$ are the predictor variables,
- $\beta_0$ is the intercept,
- $\beta_1, \beta_2, \ldots, \beta_p$ are the regression coefficients,
- $p$ is the number of predictors,
- $\varepsilon$ is the error term.

Each coefficient represents the expected change in the target variable associated with a one-unit increase in the corresponding predictor while holding all other predictors constant.

###Example:
Suppose we aim to predict the Human Development Index (HDI) of a country.

Instead of using only GDP per capita, we may consider several indicators simultaneously, such as:

- GDP per capita,
- Unemployment rate,
- Internet usage,
- Life expectancy,
- Government expenditure on education.

Multiple Linear Regression allows us to estimate the contribution of each factor while accounting for the effects of the others.

## Polynomial Regression

Simple and multiple linear regression assume that the relationship between predictors and the target variable is linear. However, many real-world relationships exhibit curved patterns that cannot be adequately represented by a straight line.

Polynomial Regression addresses this limitation by introducing polynomial terms of the predictor variables.

For example, instead of modeling the relationship using only $x$, the model may include terms such as $x^2$, $x^3$, and higher-order powers.

The polynomial regression model is defined as:

$$
y = \beta_0 + \beta_1x + \beta_2x^2 + \cdots + \beta_dx^d + \varepsilon
$$

where:

- $y$ is the dependent variable,
- $x$ is the predictor variable,
- $d$ is the degree of the polynomial,
- $\beta_0, \beta_1, \ldots, \beta_d$ are the regression coefficients,
- $\varepsilon$ is the error term.

Although the relationship between $x$ and $y$ may be nonlinear, the model remains linear with respect to its coefficients. Therefore, Polynomial Regression is still considered part of the Linear Regression Family.


###Example:
In socioeconomic applications, the relationship between GDP per capita and human development may not always be strictly linear.

At lower income levels, increases in GDP per capita may lead to substantial improvements in human development. However, beyond a certain point, additional economic growth may yield smaller gains.

Polynomial Regression can capture such nonlinear patterns more effectively than a simple straight line.

Key Idea
Polynomial Regression extends Linear Regression by introducing polynomial features while preserving linearity in the model coefficients.

This allows the model to capture nonlinear relationships without abandoning the mathematical framework of linear regression.

## Regularization

As regression models become more complex, they may begin to fit not only the underlying relationships in the data but also random noise.

This phenomenon is known as overfitting.

Regularization is a technique used to reduce model complexity and improve generalization by penalizing large coefficient values.

Instead of minimizing only the prediction error, regularized regression models also impose constraints on the magnitude of the regression coefficients.

The objective is to build models that perform well not only on the training data but also on unseen observations.

## Ridge Regression

Ridge Regression extends ordinary least squares by adding a penalty proportional to the squared magnitude of the coefficients.

The objective function is defined as:

##$$
RSS + \lambda \sum_{j=1}^{p}\beta_j^2
$$

where:

- $RSS$ is the Residual Sum of Squares,
- $\lambda$ is the regularization parameter,
- $\beta_j$ represents the coefficient associated with predictor $j$,
- $p$ is the number of predictors.

As $\lambda$ increases, the coefficients shrink toward zero.

However, Ridge Regression does not force coefficients to become exactly zero.

image.png

*Figure 3.14. Ridge Regression shrinks coefficients toward zero while retaining all predictors.*

**Note:** Ordinary Least Squares (OLS) refers to the standard Linear Regression model without regularization.

## Lasso Regression

Lasso Regression introduces a penalty based on the absolute values of the coefficients.

The objective function is defined as:

##$$
RSS + \lambda \sum_{j=1}^{p}|\beta_j|
$$

where:

- $RSS$ is the Residual Sum of Squares,
- $\lambda$ controls the strength of regularization,
- $\beta_j$ represents the coefficient associated with predictor $j$.

Unlike Ridge Regression, Lasso Regression can shrink some coefficients exactly to zero.

As a result, Lasso performs automatic feature selection.

image.png

*Figure 3.15. Lasso Regression may shrink some coefficients exactly to zero, effectively removing predictors from the model.*

## Elastic Net Regression

Elastic Net Regression combines the penalties used in Ridge and Lasso Regression.

Its objective function is defined as:

##$$
RSS
+
\lambda_1
\sum_{j=1}^{p}
|\beta_j|
+
\lambda_2
\sum_{j=1}^{p}
\beta_j^2
$$

where:

- $\lambda_1$ controls the Lasso penalty,
- $\lambda_2$ controls the Ridge penalty.

Elastic Net is particularly useful when predictors are highly correlated and feature selection is desirable.

It combines the strengths of both Ridge and Lasso Regression.

###Key Differences
| Method | Penalty | Coefficients Become Zero? | Feature Selection |
|----------|----------|----------------------------|-------------------|
| Ordinary Least Squares | None | No | No |
| Ridge Regression | $L_2$ | No | No |
| Lasso Regression | $L_1$ | Yes | Yes |
| Elastic Net Regression | $L_1 + L_2$ | Some | Yes |

*Table 3.1. Key Differences Among Linear Regression Methods.*



####Key Idea
Regularization improves model generalization by discouraging overly complex models.

Ridge Regression reduces coefficient magnitude, Lasso Regression performs feature selection, and Elastic Net combines both strategies to achieve a balance between stability and sparsity.

```python
master_df_imputed = master_df.copy()

numeric_columns = master_df_imputed.select_dtypes(
    include=["number"]
).columns

for col in numeric_columns:

    if col not in ["Human Development Index", "Year"]:

        master_df_imputed[col] = (
            master_df_imputed
            .groupby("Year")[col]
            .transform(lambda x: x.fillna(x.median()))
        )

        master_df_imputed[col] = (
            master_df_imputed[col]
            .fillna(master_df_imputed[col].median())
        )
```

**Observation:** The Lasso model retained 11 of the 13 predictors while automatically excluding two variables, thereby producing a more parsimonious and interpretable model without substantial loss of predictive accuracy.

####Model Evaluation

###Elastic Net Regression

Investigates whether combining L1 and L2 regularization can simultaneously enhance predictive performance and model interpretability.

####Hyperparameter Tuning

Identifies the optimal combination of the regularization strength (α) and the mixing parameter (l1_ratio) using cross-validation.

```python
from sklearn.linear_model import ElasticNet

elastic = ElasticNet(max_iter=10000)

param_grid = {
    "alpha": [0.0001, 0.001, 0.01, 0.1, 1],
    "l1_ratio": [0.2, 0.5, 0.8]
}

elastic_grid = GridSearchCV(
    estimator=elastic,
    param_grid=param_grid,
    scoring="neg_mean_squared_error",
    cv=cv,
    n_jobs=-1
)

elastic_grid.fit(X_train_scaled, y_train)
```

**Observation:** Cross-validation was employed to identify the optimal balance between L1 and L2 regularization while minimizing prediction error.

####Selecting the Optimal Hyperparameters

```python
print("Best Alpha:", elastic_grid.best_params_["alpha"])
print("Best L1 Ratio:", elastic_grid.best_params_["l1_ratio"])
print("Best CV Score:", np.sqrt(-elastic_grid.best_score_))
```

**Observation:** Cross-validation identified α = 0.001 and l1_ratio = 0.5 as the optimal combination of regularization parameters, indicating that a balanced contribution of L1 and L2 penalties provided the best predictive performance.

####Model Development

Estimates the Elastic Net model using the optimal hyperparameters obtained from cross-validation.

```python
elastic_model = elastic_grid.best_estimator_

y_pred_elastic = elastic_model.predict(X_test_scaled)
```

SHAP (SHapley Additive exPlanations) was employed to enhance the interpretability of the final regression model. This approach quantifies the contribution of each explanatory variable to the model predictions based on concepts derived from cooperative game theory. Unlike traditional coefficient interpretation, SHAP provides both the magnitude and direction of each variable's influence on the predicted Human Development Index (HDI).
