## Section 1: Introduction to Data Science and Business Intelligence

In today's digital world, organizations generate massive amounts of data through transactions, websites, mobile applications, sensors, and social media platforms. However, raw data has little value unless it can be transformed into meaningful information that supports decision-making and problem-solving.

Data Science, Data Analytics, and Business Intelligence are closely related disciplines that help organizations extract value from data. Although these fields share common goals, each has a distinct focus and contributes differently to organizational success.

### What is Data Science?

Data Science is an interdisciplinary field that combines statistics, mathematics, programming, and domain expertise to extract knowledge and insights from data.

The primary objective of Data Science is to transform raw data into actionable insights that support decision-making, prediction, and innovation.

Data scientists use scientific methods and computational techniques to identify patterns, build predictive models, and solve complex problems.

Common activities in Data Science include:

- Collecting data from multiple sources.
- Cleaning and preparing data for analysis.
- Exploring and visualizing data.
- Building analytical and predictive models.
- Evaluating model performance.
- Communicating findings to stakeholders.

Examples of Data Science applications include:

- Fraud detection in banking.
- Recommendation systems in e-commerce platforms.
- Customer churn prediction.
- Medical diagnosis support.
- Astronomical data analysis.

### What is Data Analytics?

Data Analytics is the process of examining datasets to discover patterns, trends, and relationships that can support decision-making.

It focuses on answering specific questions using existing data and helping organizations better understand their operations and performance.

Data Analytics is commonly used to:

- Monitor business performance.
- Identify trends and opportunities.
- Understand customer behavior.
- Investigate the causes of business problems.
- Support evidence-based decisions.

Examples include:

- Analyzing monthly sales performance.
- Evaluating marketing campaign effectiveness.
- Identifying factors contributing to customer attrition.
- Examining website traffic patterns.

### What is Business Intelligence?

Business Intelligence (BI) refers to the processes, technologies, and tools used to convert business data into meaningful information that supports managerial decision-making.

Business Intelligence primarily focuses on reporting, monitoring, and understanding the current state of the organization.

BI systems help managers answer questions such as:

- How is the business performing?
- Which products generate the highest revenue?
- Which departments require attention?
- Are organizational goals being achieved?

Common Business Intelligence tools include:

- Dashboards.
- Reports.
- Scorecards.
- Interactive visualizations.
- Performance monitoring systems.

Examples of Business Intelligence applications include:

- Executive dashboards.
- Financial reporting systems.
- Inventory monitoring solutions.
- Sales performance reports.

### Relationship Between Data Science, Data Analytics, and Business Intelligence

Although Data Science, Data Analytics, and Business Intelligence overlap in several areas, they differ in their primary objectives.

Business Intelligence focuses on monitoring and reporting historical and current business performance.

Data Analytics focuses on examining data to understand patterns, explain outcomes, and support decision-making.

Data Science extends beyond analysis by building predictive models and developing intelligent solutions capable of forecasting future outcomes.

In practice, organizations often integrate all three disciplines to maximize the value derived from data.

### Types of Data Analytics

Descriptive Analytics focuses on summarizing historical data to understand what has happened in the past.

> "What happened?"

> "What is likely to happen?"

It answers the question:

> "What should we do?"

Examples include:

Prescriptive Analytics represents the most advanced stage of analytics because it supports decision-making through actionable recommendations.

Data refers to raw facts, observations, measurements, or recordings collected about entities, events, or processes.

Examples of data include:

- Customer purchases,
- Website visits,
- Survey responses,
- Patient temperatures,
- Sensor measurements.

Without context, data often has limited meaning.

For example, consider the following observations:

120, 145, 138, 160

### From Data to Wisdom: The DIKW Hierarchy

The relationship between data and decision-making is often explained using the DIKW hierarchy.

#### Data
Raw facts without interpretation.

Example:
120, 145, 138, 160

#### Information
Data that has been processed to answer a question.

Example:
"The average daily sales are 140.75 units."

#### Knowledge
Insights gained by interpreting information.

Example:
"Daily sales appear relatively stable, allowing managers to estimate inventory requirements."

#### Wisdom
The ability to make sound decisions using knowledge and experience.

Example:
"Increasing inventory before seasonal peaks to avoid stock shortages."

The value of analytics lies in transforming raw data into actionable knowledge.

### Types of Variables

Variables can generally be classified into two categories.

Qualitative variables describe characteristics or group memberships.

Examples:

- Gender,
- Country,
- Product category.

##### Nominal Variables

Categories without any natural order.

Examples:

- Blood type,
- Nationality,
- Eye color.

##### Ordinal Variables

Categories with meaningful order.

Examples:

- Satisfaction ratings,
- Education levels,
- Risk classifications.

Although categories can be ranked, the differences between them cannot be measured numerically.



#### Quantitative (Numerical) Variables

Quantitative variables represent measurable quantities.

Examples:

- Age,
- Income,
- Revenue.

##### Discrete Variables

Countable numerical values.

Examples:

- Number of purchases,
- Number of employees.

##### Continuous Variables

Measured values that can take any value within a range.

Examples:

- Height,
- Weight,
- Temperature,
- Sales revenue.

### Levels of Measurement

According to Stevens (1946), variables can be measured using four different scales.

#### Nominal Scale

Characteristics:
- Labels only,
- No ordering.

Examples:
- Gender,
- Product category.



#### Ordinal Scale

Characteristics:
- Ordered categories,
- Unequal distances between categories.

Examples:
- Satisfaction levels,
- Credit ratings.



#### Interval Scale

Characteristics:
- Equal intervals,
- No true zero point.

Examples:
- Temperature measured in Celsius,
- IQ scores.

For example, 20°C is not twice as hot as 10°C.



#### Ratio Scale

Characteristics:
- Equal intervals,
- Presence of an absolute zero.

Examples:
- Age,
- Weight,
- Sales,
- Income.

Ratio variables support all arithmetic operations and meaningful ratio comparisons.

### Population and Sample

#### Population

The complete set of entities of interest.

Example:

All customers of a retail company.

#### Sample

A subset selected from the population.

Example:

5,000 customers chosen for analysis.

Most analytical conclusions are based on samples rather than complete populations.

## Section 3: Data Quality Issues

Data quality is one of the most critical factors affecting the success of analytical projects. Even sophisticated models and advanced analytical techniques cannot compensate for poor-quality data.

This section introduces common data quality problems encountered in practice, explains their causes and consequences, and discusses basic techniques used to identify them.

### Why This Topic Matters

A common misconception is that the majority of a data scientist's time is spent building machine learning models. In reality, a significant portion of analytical work is dedicated to understanding, validating, and preparing data.

Poor-quality data can lead to misleading insights, biased conclusions, and unreliable predictions.

This principle is often summarized by the phrase:

> **Garbage In, Garbage Out (GIGO).**

If the input data is flawed, the resulting analysis will likely be flawed as well.

Therefore, assessing data quality is an essential step before conducting any statistical analysis or developing predictive models.

### What Is Data Quality?

Data quality refers to the degree to which data is suitable for its intended purpose.

High-quality data should accurately represent the underlying phenomenon and support reliable decision-making.

Poor-quality data, on the other hand, may contain errors, inconsistencies, missing information, or biases that compromise analytical outcomes.

Data quality is not an absolute concept. Rather, it depends on the context and objectives of the analysis.

For example, a small number of missing values may be acceptable in one application but unacceptable in another, such as medical diagnosis.

### Common Data Quality Problems

#### Missing Values

**Definition**

Missing values occur when information that should be present is absent.

**Common Causes**

- Human error during data entry.
- Equipment malfunction.
- Survey non-response.
- Data integration issues.

**Potential Impact**

- Reduced statistical power.
- Biased estimates.
- Decreased model performance.
- Loss of valuable information.

**Detection Methods**

- Visual inspection.
- Missing value counts.
- Percentage of missing observations.
- Missingness heatmaps.

**Example**

A customer's age is recorded as empty because it was not provided during registration.

#### Duplicate Records

**Definition**

Duplicate records occur when the same observation appears multiple times within a dataset.

**Common Causes**

- Multiple system imports.
- Data integration errors.
- Repeated submissions.
- Lack of unique identifiers.

**Potential Impact**

- Inflated counts.
- Biased analyses.
- Distorted statistical summaries.
- Misleading business insights.

**Detection Methods**

- Duplicate checks.
- Unique identifier validation.
- Frequency analysis.

**Example**

The same customer transaction is imported twice into the database.

#### Inconsistent Data

**Definition**

Inconsistent data occurs when the same information is represented using different formats or conventions.

**Common Causes**

- Multiple data entry standards.
- Manual input variations.
- Integration from different systems.

**Potential Impact**

- Incorrect aggregations.
- Misclassification.
- Increased preprocessing complexity.

**Detection Methods**

- Frequency tables.
- Category inspection.
- Rule-based validation.

**Example**

The values "Male", "male", and "M" all represent the same category.

#### Noisy Data

**Definition**

Noisy data contains random errors, distortions, or variability that obscure the true signal.

**Common Causes**

- Sensor inaccuracies.
- Transmission errors.
- Human mistakes.
- Measurement limitations.

**Potential Impact**

- Reduced analytical accuracy.
- Increased uncertainty.
- Difficulty identifying patterns.

**Detection Methods**

- Visualization techniques.
- Statistical smoothing.
- Domain knowledge assessment.

**Example**

Temperature sensors occasionally producing abnormal fluctuations.

#### Outliers

**Definition**

Outliers are observations that differ substantially from the majority of the data.

**Common Causes**

- Data entry mistakes.
- Measurement errors.
- Genuine rare events.
- Fraudulent activity.

**Potential Impact**

- Distorted means and variances.
- Misleading statistical conclusions.
- Reduced model stability.

**Detection Methods**

- Boxplots.
- Z-score method.
- Interquartile Range (IQR) method.
- Scatter plots.

**Example**

A customer's recorded age is 250 years.

#### Redundant Features

**Definition**

Redundant features provide overlapping or duplicated information.

**Common Causes**

- Derived variables.
- Feature duplication.
- Merging multiple datasets.

**Potential Impact**

- Increased computational cost.
- Multicollinearity.
- Reduced model interpretability.

**Detection Methods**

- Correlation analysis.
- Variance Inflation Factor (VIF).
- Feature importance analysis.

**Example**

Both "Age in Years" and "Age in Months" are included as predictors.

#### Data Entry Errors

**Definition**

Data entry errors arise when information is recorded incorrectly.

**Common Causes**

- Typographical mistakes.
- Manual transcription errors.
- Incorrect coding.

**Potential Impact**

- Invalid observations.
- Misleading analyses.
- Reduced trust in the dataset.

**Detection Methods**

- Range checks.
- Validation rules.
- Logical consistency checks.

**Example**

Recording a monthly salary as 50,000 instead of 5,000.

#### Imbalanced Data (Introduction)

**Definition**

Imbalanced data occurs when one class is represented much more frequently than another.

**Why It Matters**

Models trained on imbalanced datasets may perform well overall while failing to identify minority classes effectively.

This issue is particularly important in domains such as:

- Fraud detection,
- Disease diagnosis,
- Credit risk assessment.

Techniques for handling imbalanced data will be discussed in later chapters.

#### Data Bias (Introduction)

**Definition**

Data bias refers to systematic distortions that cause data to misrepresent reality.

**Why It Matters**

Bias can lead to unfair, inaccurate, or misleading outcomes.

Examples include:

- Sampling bias,
- Selection bias,
- Measurement bias.

Understanding and mitigating bias has become an important topic in responsible data science and AI.

This topic will be revisited in greater detail in later chapters.

##Section 4 : Primary Data Cleansing

After identifying data quality issues, the next step is to address them appropriately. Data cleaning aims to improve the quality, consistency, and usability of datasets before conducting analyses or building predictive models.

Effective preprocessing enhances the reliability of analytical outcomes and reduces the risk of drawing misleading conclusions from flawed data.

### Why This Topic Matters

Real-world datasets are rarely clean. Missing values, inconsistencies, duplicates, and unusual observations are common challenges faced by analysts.

Data cleaning is often one of the most time-consuming stages of the analytical workflow. Although it may appear less exciting than model development, it is essential for producing valid and trustworthy results.

Importantly, there is no universal cleaning strategy. The most appropriate preprocessing techniques depend on the characteristics of the data and the objectives of the analysis.

### Handling Missing Values

Missing values occur when information is unavailable for certain observations. Several strategies can be used to address missingness, each with advantages and limitations.

#### Removal

Rows or columns containing missing values are removed from the dataset.

**Advantages**
- Simple to implement.
- Preserves original values.

**Limitations**
- May lead to information loss.
- Can introduce bias if missingness is not random.

#### Mean Imputation

Missing numerical values are replaced by the arithmetic mean.

**Advantages**
- Easy to apply.
- Maintains sample size.

**Limitations**
- Sensitive to outliers.
- Reduces variability.

#### Median Imputation

Missing numerical values are replaced by the median.

**Advantages**
- Robust to outliers.
- Maintains sample size.

**Limitations**
- May distort distributions.

#### Mode Imputation

Missing categorical values are replaced using the most frequent category.

**Advantages**
- Simple and intuitive.
- Suitable for categorical variables.

**Limitations**
- May overrepresent dominant categories.

#### Interpolation

Missing values are estimated using surrounding observations.

**Advantages**
- Useful for ordered or time-series data.
- Preserves trends.

**Limitations**
- Not appropriate for all datasets.

```python
# Removal
df_drop = df_quality.dropna()

print("After Removing Missing Values:")
display(df_drop)
```

```python
# Mean Imputation
df_mean = df_quality.copy()

df_mean["Age"] = (
    df_mean["Age"]
    .fillna(df_mean["Age"].mean())
)

print("Mean Imputation:")
display(df_mean)
```

```python
# Median Imputation
df_median = df_quality.copy()

df_median["Age"] = (
    df_median["Age"]
    .fillna(df_median["Age"].median())
)

print("Median Imputation:")
display(df_median)
```

```python
# Mode Imputation
df_mode = df_quality.copy()

df_mode["Gender"] = (
    df_mode["Gender"]
    .fillna(df_mode["Gender"].mode()[0])
)

print("Mode Imputation:")
display(df_mode)
```

```python
# Interpolation
df_interp = df_quality.copy()

df_interp["Sales"] = (
    df_interp["Sales"]
    .interpolate()
)

print("Interpolation:")
display(df_interp)
```

```python
df_clean = df_quality.copy()

# Numerical variables
df_clean["Age"] = (
    df_clean["Age"]
    .fillna(df_clean["Age"].median())
)

# Categorical variables
df_clean["Gender"] = (
    df_clean["Gender"]
    .fillna(df_clean["Gender"].mode()[0])
)

# Sales interpolation
df_clean["Sales"] = (
    df_clean["Sales"]
    .interpolate()
)

print("Cleaned Dataset:")
display(df_clean)

print("\nRemaining Missing Values:")
print(df_clean.isnull().sum())
```

### Interpreting the Results

The results show that different imputation techniques can produce substantially different outcomes. Mean imputation was sensitive to the extreme age value, whereas median imputation provided a more robust estimate. Mode imputation was suitable for the categorical variable, and interpolation effectively estimated the missing sales value. These findings highlight that data cleaning decisions should be guided by the nature of the data and the analytical objective.

### Key Insights

- Removing missing values is simple but may result in information loss.
- Mean imputation is sensitive to extreme observations.
- Median imputation is more robust in the presence of outliers.
- Mode imputation is appropriate for categorical variables.
- Interpolation is useful when data follow a natural sequence.
- Real-world preprocessing often combines several cleaning strategies to produce an analysis-ready dataset.

##Handling Duplicate Values

After identifying duplicate entities, analysts must determine the most appropriate strategy for resolving them. Depending on the analytical context, duplicate records may be removed, merged, or retained if they provide complementary information.

```python
df_no_duplicates = (
    df_quality
    .drop_duplicates(
        subset="Customer_ID",
        keep="first"
    )
)

print("After Removing Duplicate Customer IDs:")

display(df_no_duplicates)
```

The duplicate records associated with `Customer_ID` 1002 were consolidated by retaining only the first occurrence. This example illustrates that duplicate handling requires analytical judgment, as the most complete or reliable record may not always be the first one.

##Handling Inconsistent Data

Categorical variables often contain multiple representations of the same concept. Standardizing these values improves consistency and supports more reliable analyses.

```python
gender_mapping = {
    "M": "Male",
    "male": "Male",
    "Female": "Female",
    "female": "Female"
}

df_standardized = df_quality.copy()

df_standardized["Gender"] = (
    df_standardized["Gender"]
    .replace(gender_mapping)
)

print("Standardized Categories:")

print(df_standardized["Gender"].value_counts(dropna=False))
```

The standardization process successfully unified inconsistent representations of the `Gender` variable. This simple preprocessing step improves data consistency and reduces the likelihood of misleading analytical outcomes.

##Noise Reduction

Noise reduction techniques help minimize random fluctuations and improve the interpretability of data.

###Smoothing

## Measures of Dispersion

While measures of central tendency describe where observations tend to cluster, measures of dispersion quantify the extent to which observations vary around their central values.

The most commonly used measures of dispersion include:

### Variance

Variance measures the average squared deviation of observations from their mean. Larger variance values indicate greater variability within the dataset.

$$

##Statistical Inference

Descriptive statistics summarize the characteristics of observed data, whereas statistical inference allows analysts to draw conclusions about a larger population based on sample information.

Inferential methods help determine whether observed patterns are likely to reflect genuine population characteristics or simply arise from random variation.

The fundamental concepts include:

- Hypothesis Testing
- Significance Level
- P-value
- Confidence Intervals

### Hypothesis Testing

Hypothesis testing is a formal statistical procedure used to evaluate claims about a population using sample data.

Two competing hypotheses are formulated:

- **Null Hypothesis (H₀):** Assumes no effect, no difference, or no relationship.
- **Alternative Hypothesis (H₁):** Assumes an effect, difference, or relationship exists.

The objective is to determine whether the available evidence is sufficiently strong to reject the null hypothesis.

### Significance Level (α)

The significance level represents the maximum probability of committing a Type I error that researchers are willing to tolerate.

A Type I error occurs when the null hypothesis is rejected even though it is actually true.

In practice, the most commonly used significance level is:

$$
\alpha = 0.05
$$

This implies accepting a 5% risk of falsely rejecting the null hypothesis.

### P-value

The p-value quantifies the probability of obtaining results at least as extreme as those observed, assuming that the null hypothesis is true.

The p-value does not indicate the probability that the null hypothesis is true or false. Instead, it measures the compatibility of the observed data with the null hypothesis.

### Confidence Intervals

A confidence interval provides a range of plausible values for an unknown population parameter.

Rather than relying solely on a single point estimate, confidence intervals quantify the uncertainty associated with sample estimates.

The most commonly reported confidence level is 95%.

A 95% confidence interval means that if the same sampling procedure were repeated many times, approximately 95% of the resulting intervals would contain the true population parameter.

### Demonstration

The following example estimates a 95% confidence interval for the average salary.

```python
from scipy.stats import t
import numpy as np

salary = eda_df["Salary"]

mean_salary = salary.mean()
std_salary = salary.std(ddof=1)
n = len(salary)

confidence_level = 0.95
alpha = 1 - confidence_level

t_critical = t.ppf(
    1 - alpha / 2,
    df=n - 1
)

margin_of_error = (
    t_critical * std_salary / np.sqrt(n)
)

lower_bound = mean_salary - margin_of_error
upper_bound = mean_salary + margin_of_error

print("Sample Mean:", round(mean_salary, 2))
print("95% Confidence Interval:")
print(
    f"({lower_bound:.2f}, {upper_bound:.2f})"
)
```

### Interpreting the Results

Statistical inference extends beyond describing observed data by enabling conclusions about broader populations. Hypothesis testing provides a structured framework for evaluating claims, while significance levels and p-values support evidence-based decision-making.

Confidence intervals complement hypothesis tests by quantifying the uncertainty surrounding sample estimates and presenting a range of plausible values rather than a single estimate.

### Common Misconceptions

- A p-value is not the probability that the null hypothesis is true.
- Failing to reject the null hypothesis does not prove that it is true.
- Statistical significance does not necessarily imply practical significance.
- Confidence intervals describe uncertainty in estimation rather than certainty about outcomes.

### Summary of Inferential Concepts

| Concept | Purpose |
|---|---|
| Null Hypothesis (H₀) | Represents the default assumption |
| Alternative Hypothesis (H₁) | Represents the competing claim |
| Significance Level (α) | Defines the acceptable Type I error rate |
| P-value | Measures compatibility with H₀ |
| Confidence Interval | Quantifies estimation uncertainty |

##Statistical Tests

Statistical tests are inferential procedures used to evaluate hypotheses and determine whether observed differences or relationships are statistically significant.

The selection of an appropriate statistical test depends on the research objective, the type of variables involved, and the assumptions underlying each method.

This section introduces three widely used statistical tests:

- Independent Samples T-Test
- Chi-Square Test of Independence
- One-Way ANOVA

### Independent Samples T-Test

The independent samples t-test is used to determine whether the means of two independent groups differ significantly.

Example question:

> Do male and female employees receive different average salaries?

#### Assumptions

- The dependent variable is continuous.
- The two groups are independent.
- Observations are independent.
- The dependent variable is approximately normally distributed within each group.
- Homogeneity of variance is assumed.

#### Hypotheses

Null Hypothesis (H₀):

The average salaries of male and female employees are equal.

$$
H_0:\mu_{Male}=\mu_{Female}
$$

Alternative Hypothesis (H₁):

The average salaries of male and female employees differ.

$$
H_1:\mu_{Male}\neq\mu_{Female}
$$

```python
from scipy.stats import ttest_ind

male_salary = eda_df.loc[
    eda_df["Gender"] == "Male",
    "Salary"
]

female_salary = eda_df.loc[
    eda_df["Gender"] == "Female",
    "Salary"
]

t_statistic, p_value = ttest_ind(
    male_salary,
    female_salary
)

print("T-statistic:", round(t_statistic, 4))
print("P-value:", round(p_value, 4))
```

If the p-value is less than or equal to 0.05, the null hypothesis is rejected, indicating a statistically significant difference between the average salaries of the two groups.

Otherwise, there is insufficient evidence to conclude that a significant difference exists.

### Chi-Square Test of Independence

The chi-square test evaluates whether two categorical variables are associated.

Example question:

> Is employee gender associated with department assignment?

#### Assumptions

- Variables are categorical.
- Observations are independent.
- Expected frequencies are sufficiently large.

#### Hypotheses

Null Hypothesis (H₀):

Gender and department are independent.

Alternative Hypothesis (H₁):

Gender and department are associated.

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(
    eda_df["Gender"],
    eda_df["Department"]
)

display(contingency_table)

chi2, p_value, dof, expected = chi2_contingency(
    contingency_table
)

print("Chi-Square Statistic:", round(chi2, 4))
print("P-value:", round(p_value, 4))
print("Degrees of Freedom:", dof)
```

A p-value less than or equal to 0.05 suggests that the two categorical variables are statistically associated.

Otherwise, there is insufficient evidence to conclude that an association exists.

### One-Way ANOVA

One-Way Analysis of Variance (ANOVA) is used to compare the means of three or more independent groups.

Example question:

> Do average salaries differ across departments?

#### Assumptions

- The dependent variable is continuous.
- The groups are independent.
- Observations are independent.
- The dependent variable is approximately normally distributed within groups.
- Homogeneity of variances is assumed.

#### Hypotheses

Null Hypothesis (H₀):

All group means are equal.

$$
H_0:\mu_1=\mu_2=\cdots=\mu_k
$$

Alternative Hypothesis (H₁):

At least one group mean differs.

```python
from scipy.stats import f_oneway

hr_salary = eda_df.loc[
    eda_df["Department"] == "HR",
    "Salary"
]

it_salary = eda_df.loc[
    eda_df["Department"] == "IT",
    "Salary"
]

finance_salary = eda_df.loc[
    eda_df["Department"] == "Finance",
    "Salary"
]

f_statistic, p_value = f_oneway(
    hr_salary,
    it_salary,
    finance_salary
)

print("F-statistic:", round(f_statistic, 4))
print("P-value:", round(p_value, 4))
```

If the p-value is less than or equal to 0.05, the null hypothesis is rejected, indicating that at least one group mean differs significantly.

However, ANOVA does not identify which specific groups differ. Post-hoc tests are required for detailed pairwise comparisons.

### Choosing the Appropriate Statistical Test

| Research Objective | Variable Types | Recommended Test |
|---|---|---|
| Compare two group means | Numerical vs Binary Categorical | Independent T-Test |
| Compare three or more group means | Numerical vs Multi-category Categorical | One-Way ANOVA |
| Assess association between categories | Categorical vs Categorical | Chi-Square Test |

### Interpreting the Results

Statistical tests transform exploratory observations into evidence-based conclusions. While descriptive analyses may suggest differences or associations, inferential tests determine whether these findings are statistically significant or likely attributable to random variation.

The choice of statistical test should always be guided by the analytical objective, the nature of the variables involved, and the assumptions underlying each procedure.

In analytical practice, statistical testing complements exploratory analysis by providing formal decision-making frameworks. Analysts should avoid relying exclusively on p-values and instead consider effect sizes, practical significance, and domain context when interpreting results.

**Industry Insight:** T-tests, Chi-Square tests, and ANOVA are among the most frequently used inferential tools in business analytics, A/B testing, market research, healthcare analytics, and social science studies. Demonstrating the ability to select and interpret these tests appropriately highlights strong analytical reasoning rather than merely technical implementation.

## 4.7 Relationship Analysis

### Pearson Correlation

Pearson correlation measures the strength and direction of a linear relationship between two continuous variables.

Its values range from -1 to +1:

- +1 indicates a perfect positive linear relationship.
- -1 indicates a perfect negative linear relationship.
- 0 indicates no linear relationship.

####Mathematical Formula:

$$
r=
\frac{
\sum_{i=1}^{n}(x_i-\bar{x})(y_i-\bar{y})
}{
\sqrt{\sum(x_i-\bar{x})^2}
\sqrt{\sum(y_i-\bar{y})^2}
}
$$

```python
pearson_corr = eda_df["Experience"].corr(
    eda_df["Salary"],
    method="pearson"
)

print(
    "Pearson Correlation:",
    round(pearson_corr, 3)
)
```

###Interpretation Guide

|---|---|

| 0.20 – 0.39 | Weak |

| 0.60 – 0.79 | Strong |
| 0.80 – 1.00 | Very strong |

### Spearman Correlation

Spearman correlation evaluates the strength and direction of a monotonic relationship using ranked observations.

Unlike Pearson correlation, it does not assume linearity and is less sensitive to outliers.

```python
spearman_corr = eda_df["Experience"].corr(
    eda_df["Salary"],
    method="spearman"
)

print(
    "Spearman Correlation:",
    round(spearman_corr, 3)
)
```

####When to Use Spearman

- Outliers may influence Pearson correlation.

### Simple Linear Regression

Simple linear regression models the relationship between a predictor variable and an outcome variable by fitting a straight line to the observed data.

Example question:

> Can years of experience help explain salary differences?

####Regression Equation:
$$
y=\beta_0+\beta_1x+\varepsilon
$$

where:

- $y$ is the dependent variable,
- $x$ is the independent variable,
- $\beta_0$ is the intercept,
- $\beta_1$ is the slope,
- $\varepsilon$ is the error term.

```python
from scipy.stats import linregress

regression = linregress(
    eda_df["Experience"],
    eda_df["Salary"]
)

print("Slope:", round(regression.slope, 2))
print("Intercept:", round(regression.intercept, 2))
print("R-squared:", round(regression.rvalue**2, 3))
print("P-value:", round(regression.pvalue, 4))
```

### Interpreting Regression Output

- The slope represents the expected change in salary associated with a one-unit increase in experience.
- The intercept represents the predicted salary when experience equals zero.
- The p-value evaluates whether the observed relationship is statistically significant.
- R² indicates the proportion of variability in salary explained by experience.

### Multiple Linear Regression

Multiple linear regression extends simple linear regression by examining the simultaneous effect of multiple predictor variables on a single continuous outcome variable.

Example question:

> To what extent do age and experience jointly explain salary differences?

####Regression Equation:
$$
y=\beta_0+\beta_1x_1+\beta_2x_2+\cdots+\beta_kx_k+\varepsilon
$$

where:

- $y$ is the dependent variable,
- $\beta_0$ is the intercept,
- $\beta_1,\beta_2,\ldots,\beta_k$ are regression coefficients,
- $x_1,x_2,\ldots,x_k$ are predictor variables,
- $\varepsilon$ is the error term.

```python
import statsmodels.api as sm

X = eda_df[["Age", "Experience"]]
y = eda_df["Salary"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```

### Interpreting the Results

The regression output provides several important statistics:

- The intercept represents the predicted salary when all predictors equal zero.
- Each regression coefficient estimates the expected change in salary associated with a one-unit increase in the corresponding predictor, holding other variables constant.
- The p-values assess whether individual predictors contribute significantly to the model.
- The R² statistic indicates the proportion of salary variability explained collectively by the predictors.
- The adjusted R² accounts for the number of predictors included in the model and provides a more reliable measure of explanatory power.

### Comparing Regression Approaches

| Method | Number of Predictors | Primary Objective |
|---|---:|---|
| Simple Linear Regression | 1 | Examine the effect of a single predictor |
| Multiple Linear Regression | 2 or more | Evaluate the joint contribution of multiple predictors |

### Coefficient of Determination (R²)

The coefficient of determination quantifies the proportion of variation in the dependent variable that can be explained by the independent variable.

Its values range from 0 to 1.

Higher values indicate greater explanatory power.

####Interpretation Guide

| R² Value | Interpretation |
|---|---|
| Close to 0 | Weak explanatory ability |
| Around 0.50 | Moderate explanatory ability |
| Close to 1 | Strong explanatory ability |

### Adjusted R²

Unlike R², adjusted R² penalizes the inclusion of unnecessary predictors.

As additional variables are introduced into a model, R² never decreases. Adjusted R² helps determine whether the added predictors genuinely improve the model's explanatory ability.

### Interpreting the Results

Relationship analysis demonstrated how different techniques capture associations and predictive relationships between variables. Pearson correlation quantified linear relationships, whereas Spearman correlation assessed monotonic associations that may not be strictly linear.

Simple linear regression provided a predictive framework for examining how changes in a single predictor variable were associated with changes in an outcome variable. Multiple linear regression extended this framework by evaluating the simultaneous contribution of multiple predictors while controlling for the effects of other variables.

The coefficient of determination (R²) quantified the proportion of variability in the outcome variable explained by the regression model, whereas adjusted R² provided a more reliable measure by accounting for the number of predictors included.

Together, these methods provide valuable insights into association, explanation, and prediction during exploratory analysis and support a deeper understanding of the factors that influence business outcomes.

##Encoding

Many machine learning algorithms require numerical inputs and cannot directly process categorical variables. Encoding techniques transform categorical data into numerical representations while preserving the information contained within the original categories.

The choice of encoding technique depends on the nature of the categorical variable and the analytical objective.

###Original Categorical Variables

```python
print("Original Categorical Variables:")

display(
    df_preprocessing[
        ["Education", "Department", "City"]
    ]
)
```

### One-Hot Encoding

One-hot encoding Converts each category into a new binary column (0 or 1), with a column for each unique category in the original variable.

```python
from sklearn.preprocessing import OneHotEncoder
onehot_df = pd.get_dummies(
    df_preprocessing,
    columns=["Department"],
    dtype=int
)

print("One-Hot Encoding:")

display(onehot_df)
```

####Use Case: Best for nominal (unordered) categorical variables with a limited number of unique values.
####Pros: Captures categorical relationships without implying any order.
####Cons: Increases dimensionality significantly, especially with high-cardinality features (many unique categories).

### Label Encoding

Label Encoding assigns each unique category an integer. Categories are replaced with integer values, but the encoding does not imply order.

```python
from sklearn.preprocessing import LabelEncoder

label_df = df_preprocessing.copy()

label_encoder = LabelEncoder()

label_df["Department_Label"] = label_encoder.fit_transform(
    label_df["Department"]
)

print("Label Encoding:")

display(
    label_df[
        ["Department", "Department_Label"]
    ]
)
```

####Use Case: Primarily for nominal categorical variables, especially when high cardinality makes one-hot encoding impractical.
####Pros: Simple and does not increase dimensionality.
####Cons: Can mislead models that interpret integers as ordered, potentially creating a false sense of order.

###Ordinal Encoding
Ordinal Encoding assigns each category an integer while preserving the order. Typically requires an explicit ordering of the categories.

```python
from sklearn.preprocessing import OrdinalEncoder
education_order = {
    "High School": 1,
    "Bachelor": 2,
    "Master": 3,
    "PhD": 4
}

ordinal_df = df_preprocessing.copy()

ordinal_df["Education_Ordinal"] = (
    ordinal_df["Education"]
    .map(education_order)
)

print("Ordinal Encoding:")

display(
    ordinal_df[
        ["Education", "Education_Ordinal"]
    ]
)
```

####Use Case: Best for ordinal (ordered) categorical variables, where categories have a meaningful progression.
####Pros: Encodes the natural order of categories without inflating dimensionality.
####Cons: Assumes a linear relationship between categories, which may not always be appropriate.


###Target Encoding (Mean Encoding)
Target Encoding replaces each category with the mean of the target variable for that category. Often used with cross-validation to avoid data leakage.

```python
target_df = pd.DataFrame({
    "Department": ["HR", "IT", "Finance", "IT", "HR"],
    "Performance_Score": [70, 90, 80, 95, 75]
})

display(target_df)
```

####Use Case: For categorical variables with a significant relationship to the target variable (often used in regression tasks).
####Pros: Captures complex relationships between the category and the target variable.
####Cons: Prone to overfitting and data leakage if not implemented carefully.

### Frequency Encoding
Frequency Encoding replaces each category with the frequency or count of occurrences in the dataset.

```python
frequency_df = df_preprocessing.copy()

frequency_map = (
    frequency_df["City"]
    .value_counts()
)

frequency_df["City_Frequency"] = (
    frequency_df["City"]
    .map(frequency_map)
)

print("Frequency Encoding:")

display(
    frequency_df[
        ["City", "City_Frequency"]
    ]
)
```

####Use Case: Useful for high-cardinality variables to reduce dimensionality and add information about the distribution of categories.
####Pros: Reduces dimensionality while retaining some information about category prevalence.
####Cons: Does not capture any inherent relationship between the category and target variable.

##Feature Scaling

Numerical variables are often measured using different units and ranges. For example, one variable may range from 0 to 100, while another ranges from thousands to millions.

Feature scaling transforms numerical variables into comparable scales, preventing variables with larger magnitudes from dominating analytical results and improving the performance of many machine learning algorithms.

###Original Numerical Variables

```python
print("Original Numerical Variables:")

display(
    df_preprocessing[
        ["Experience", "Salary"]
    ]
)
```

### Min-Max Scaling

Min-Max Scaling transforms numerical values into a predefined range, typically between 0 and 1.

The transformation is defined as:

$$
x'=\frac{x-x_{\min}}{x_{\max}-x_{\min}}
$$


This method preserves the shape of the original distribution but is sensitive to outliers.

```python
from sklearn.preprocessing import MinMaxScaler

minmax_df = df_preprocessing.copy()

scaler = MinMaxScaler()

minmax_df[["Experience", "Salary"]] = scaler.fit_transform(
    minmax_df[["Experience", "Salary"]]
)

print("Min-Max Scaling:")

display(
    minmax_df[
        ["Experience", "Salary"]
    ]
)
```

The standardized values were centered around zero and expressed in standard deviation units.The transformed values were rescaled to the range between 0 and 1 while preserving their relative ordering.

### Z-Score Standardization

Standardization transforms variables so that they have a mean of zero and a standard deviation of one.

The transformation is defined as:

$$
z=\frac{x-\mu}{\sigma}
$$
where:

- (x) is the original value,
- (μ) is the mean,
- (σ) is the standard deviation.


Unlike Min-Max Scaling, standardization does not constrain values to a fixed interval.

```python
from sklearn.preprocessing import StandardScaler

standard_df = df_preprocessing.copy()

standard_scaler = StandardScaler()

standard_df[["Experience", "Salary"]] = standard_scaler.fit_transform(
    standard_df[["Experience", "Salary"]]
)

print("Z-Score Standardization:")

display(
    standard_df[
        ["Experience", "Salary"]
    ]
)
```

The standardized values were centered around zero and expressed in standard deviation units.

### Robust Scaling

Robust Scaling transforms numerical variables using the median and the interquartile range (IQR) rather than the mean and standard deviation.

Unlike Min-Max Scaling and Standardization, this approach is less sensitive to extreme observations and is particularly useful when datasets contain outliers.

The transformation is defined as:

$$
x'=\frac{x-\mathrm{Median}(x)}{IQR}
$$

where:

- Median is the 50th percentile of the variable.
- IQR (Interquartile Range) is calculated as:

$$
IQR = Q_3 - Q_1
$$

```python
from sklearn.preprocessing import RobustScaler

robust_df = df_preprocessing.copy()

robust_scaler = RobustScaler()

robust_df[["Experience", "Salary"]] = robust_scaler.fit_transform(
    robust_df[["Experience", "Salary"]]
)

print("Robust Scaling:")

display(
    robust_df[
        ["Experience", "Salary"]
    ]
)
```

The transformed values were less influenced by extreme observations due to the use of the median and IQR.

Robust Scaling transformed the numerical variables using the median and interquartile range. Because it is less influenced by extreme observations, it provides a more stable representation when outliers are present in the data.

### Comparing Feature Scaling Methods

- **Min-Max Scaling** rescales values to a fixed range, typically between 0 and 1, but is sensitive to outliers.
- **Z-Score Standardization** centers variables around zero with unit variance and is widely used in statistical and machine learning applications.
- **Robust Scaling** relies on the median and IQR, making it more resistant to the influence of extreme values.

The choice of scaling technique should depend on the characteristics of the data and the assumptions of the intended analytical method.

### When to Use Each Method

| Method | Best Used When | Limitation |
|---|---|---|
| Min-Max Scaling | Fixed range is required | Sensitive to outliers |
| StandardScaler | Features should have zero mean and unit variance | Influenced by outliers |
| RobustScaler | The dataset contains outliers | Does not preserve a fixed range |




##Vector Normalization

Unlike feature scaling, which adjusts the range of individual variables, normalization transforms each observation so that its overall magnitude becomes standardized.

Normalization is particularly useful in algorithms that rely on distance calculations or vector similarity, such as K-Nearest Neighbors (KNN), Support Vector Machines (SVM), and text mining applications.

###Original Numerical Variables

```python
print("Original Numerical Variables:")

display(
    df_preprocessing[
        ["Experience", "Salary"]
    ]
)
```

###L1 Normalization

L1 normalization scales each observation so that the sum of the absolute values of its features equals one.


####Mathematical Formula:

$$
x_i'=\frac{x_i}{\sum_{j=1}^{n}|x_j|}
$$

```python
from sklearn.preprocessing import Normalizer

l1_df = df_preprocessing.copy()

l1_normalizer = Normalizer(norm="l1")

l1_df[["Experience", "Salary"]] = l1_normalizer.fit_transform(
    l1_df[["Experience", "Salary"]]
)

comparison_l1 = pd.DataFrame({
    "Experience (Original)": df_preprocessing["Experience"],
    "Experience (L1)": l1_df["Experience"],
    "Salary (Original)": df_preprocessing["Salary"],
    "Salary (L1)": l1_df["Salary"]
})

print("L1 Normalization:")

display(comparison_l1)
```

### L2 Normalization

L2 normalization scales each observation so that its Euclidean length equals one.

####Mathematical Formula:
$$
x_i'=\frac{x_i}{\sqrt{\sum_{j=1}^{n}x_j^2}}
$$

```python
l2_df = df_preprocessing.copy()

l2_normalizer = Normalizer(norm="l2")

l2_df[["Experience", "Salary"]] = l2_normalizer.fit_transform(
    l2_df[["Experience", "Salary"]]
)

comparison_l2 = pd.DataFrame({
    "Experience (Original)": df_preprocessing["Experience"],
    "Experience (L2)": l2_df["Experience"],
    "Salary (Original)": df_preprocessing["Salary"],
    "Salary (L2)": l2_df["Salary"]
})

print("L2 Normalization:")

display(comparison_l2)
```

### Interpreting the Results

Unlike scaling techniques that operate on individual features, normalization transforms entire observations. L1 normalization preserves proportional contributions based on absolute magnitudes, whereas L2 normalization preserves geometric relationships by constraining observations to unit length. These techniques are particularly valuable in distance-based and similarity-based analytical methods.

### Comparing Normalization Methods

| Method | Main Idea | Common Applications |
|---|---|---|
| L1 Normalization | Sum of absolute values equals 1 | Sparse data, text mining |
| L2 Normalization | Euclidean norm equals 1 | KNN, SVM, similarity measures |

##Discretization and Binning

Discretization is the process of transforming continuous numerical variables into a finite number of intervals or categories. This technique simplifies data representation, improves interpretability, and can sometimes enhance analytical performance.

Binning is one of the most common approaches to discretization and involves grouping observations into meaningful intervals.

###Original Data

```python
age_df = pd.DataFrame({
    "Age": [18, 22, 27, 31, 36, 42, 48, 55, 61, 70]
})

print("Original Data:")

display(age_df)
```

### Manual Binning

Manual binning relies on domain knowledge to define meaningful intervals based on analytical objectives or business rules.

```python
manual_df = age_df.copy()

manual_bins = [0, 25, 40, 60, 100]

manual_labels = [
    "Young Adult",
    "Adult",
    "Middle-Aged",
    "Senior"
]

manual_df["Age_Group"] = pd.cut(
    manual_df["Age"],
    bins=manual_bins,
    labels=manual_labels
)

print("Manual Binning:")

display(manual_df)
```

The observations were grouped according to predefined business-oriented categories.

###Automatic Binning

####**Equal Width Binning**

Equal Width Binning divides the range of a variable into intervals of identical width.

####Mathematical Formula

$$
\text{Bin Width}
=
\frac{\max(x)-\min(x)}{k}
$$

where:

k represents the number of bins

```python
equal_width_df = age_df.copy()

equal_width_df["Equal_Width"] = pd.cut(
    equal_width_df["Age"],
    bins=4
)

print("Equal Width Binning:")

display(equal_width_df)
```

### Equal Frequency Binning

Equal Frequency Binning creates intervals containing approximately the same number of observations.

```python
equal_freq_df = age_df.copy()

equal_freq_df["Equal_Frequency"] = pd.qcut(
    equal_freq_df["Age"],
    q=4
)

print("Equal Frequency Binning:")

display(equal_freq_df)
```

### Interpreting the Results

Although all three approaches transformed continuous values into categories, they served different purposes. Manual binning incorporated domain knowledge and business understanding, Equal Width Binning prioritized uniform interval sizes, whereas Equal Frequency Binning emphasized balanced distributions across bins.

Therefore, the choice of binning strategy should align with both the characteristics of the data and the objectives of the analysis.

### Comparing Binning Methods

| Method | Main Idea | Advantage | Limitation |
|---|---|---|---|
| Manual Binning | Business-defined intervals | Highly interpretable | Subjective |
| Equal Width | Equal interval sizes | Simple to implement | Sensitive to skewness |
| Equal Frequency | Equal observations per bin | Balanced distributions | Unequal interval widths |

## Feature Transformation

Feature transformation refers to the application of mathematical functions to numerical variables in order to modify their distributions and improve their analytical properties.

These transformations are commonly used to reduce skewness, stabilize variance, and make variables more suitable for statistical analyses and machine learning algorithms.

###Original Data

```python
transformation_df = pd.DataFrame({
    "Income": [1000, 1500, 1800, 2200, 3000, 4500, 7000, 12000, 25000, 50000]
})

print("Original Data:")

display(transformation_df)
```

### Log Transformation

Log transformation compresses large values more strongly than small values, thereby reducing positive skewness and mitigating the influence of extreme observations.
####Mathematical Formula:

$$
x'=\log(x+1)
$$

where:

- \(x\) is the original value,
- \(x'\) is the transformed value.

```python
log_df = transformation_df.copy()

log_df["Income_Log"] = np.log1p(
    log_df["Income"]
)

comparison_log = pd.DataFrame({
    "Original Income": transformation_df["Income"],
    "Log Transformed": log_df["Income_Log"]
})

print("Log Transformation:")

display(comparison_log)
```

The transformation reduced the spacing between large income values while preserving their relative ordering.

### Square Root Transformation

####Mathematical Formula
$$
x'=\sqrt{x}
$$

Square root transformation reduces moderate positive skewness while preserving the interpretability of the original variable.

```python
sqrt_df = transformation_df.copy()

sqrt_df["Income_Sqrt"] = np.sqrt(
    sqrt_df["Income"]
)

comparison_sqrt = pd.DataFrame({
    "Original Income": transformation_df["Income"],
    "Square Root": sqrt_df["Income_Sqrt"]
})

print("Square Root Transformation:")

display(comparison_sqrt)
```

The transformation moderately compressed larger values while maintaining the original ranking of observations.

### Interpreting the Results

Both transformations altered the distribution of the income variable by reducing the influence of large values. However, log transformation produced a stronger compression effect and is generally preferred when substantial positive skewness exists. In contrast, square root transformation provides a milder adjustment and may be appropriate for moderately skewed variables.

### Comparing Transformation Methods

| Method | Primary Purpose | Suitable For |
|---|---|---|
| Log Transformation | Strong reduction of positive skewness | Highly skewed variables |
| Square Root Transformation | Moderate reduction of skewness | Mildly skewed variables |

##Data Splitting and Data Leakage Awareness

Before developing predictive models, datasets should be partitioned into separate subsets to ensure that model performance is evaluated fairly on unseen data.

Using the entire dataset for both training and evaluation may lead to overly optimistic results and poor generalization to new observations.

### Training and Test Sets

A common practice is to divide the dataset into:

- **Training Set:** Used to learn patterns and relationships from the data.
- **Test Set:** Used to evaluate the model's performance on unseen observations.

This separation provides a more realistic estimate of how the model is expected to perform in real-world applications.

###Demonstration Dataset

```python
split_df = pd.DataFrame({
    "Experience": [2, 5, 8, 12, 4, 6, 9, 15, 3, 7],
    "Salary": [2500, 4200, 6100, 8500, 3800, 5000, 6800, 9200, 3000, 5500],
    "Promotion": [0, 0, 1, 1, 0, 1, 1, 1, 0, 1]
})

print("Original Dataset:")

display(split_df)
```

###Train-Test Split

```python
from sklearn.model_selection import train_test_split

X = split_df.drop("Promotion", axis=1)
y = split_df["Promotion"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Features Shape:", X_train.shape)
print("Testing Features Shape:", X_test.shape)

print("\nTraining Labels Shape:", y_train.shape)
print("Testing Labels Shape:", y_test.shape)
```

The dataset was divided into training and testing subsets using an 80:20 ratio.

### Validation Set (Introduction)

In more advanced workflows, a validation set may also be used to tune model hyperparameters and compare alternative models before evaluating final performance on the test set.

A common workflow is:

Training Set → Validation Set → Test Set

### Data Leakage Awareness

Data leakage occurs when information from outside the training process becomes available to the model during development. As a result, the reported performance may appear unrealistically high and fail to generalize to new data.

Common examples include:

- Performing preprocessing on the entire dataset before splitting.
- Using future information to predict past events.
- Accidentally incorporating target-related information into predictor variables.

### Best Practices to Avoid Data Leakage

- Split the data before preprocessing whenever appropriate.
- Fit preprocessing techniques using only the training data.
- Apply the fitted transformations to the testing data.
- Keep the test set isolated until the final evaluation stage.

### Interpreting the Results

Data splitting is a fundamental step in predictive analytics because it enables objective model evaluation. Separating training and testing data helps estimate how well a model generalizes to unseen observations, while awareness of data leakage prevents overly optimistic and misleading performance assessments.

### Interaction Features

#### Example: Experience × Education Level

The effect of work experience on salary may depend on educational attainment.

```python
feature_df = pd.DataFrame({
    "Experience": [2, 5, 8, 10],
    "Education_Level": [12, 16, 16, 18]
})

feature_df["Experience_Education"] = (
    feature_df["Experience"]
    * feature_df["Education_Level"]
)

display(feature_df)
```

### Aggregation Features

Aggregation features summarize information across groups and provide broader contextual insights.

They are particularly useful when observations belong to meaningful categories.

#### Example: Average Department Salary

```python
feature_df = pd.DataFrame({
    "Department": ["HR", "IT", "IT", "Finance", "HR"],
    "Salary": [3500, 5000, 5200, 4800, 3700]
})

feature_df["Department_Avg_Salary"] = (
    feature_df.groupby("Department")["Salary"]
    .transform("mean")
)

display(feature_df)
```

### Date-Based Features

Date variables often contain hidden information that can be extracted into meaningful features.

Examples include year, month, day of the week, quarter, and seasonality indicators.

```python
feature_df = pd.DataFrame({
    "Order_Date": pd.to_datetime([
        "2025-01-15",
        "2025-03-22",
        "2025-07-10",
        "2025-12-05"
    ])
})

feature_df["Month"] = (
    feature_df["Order_Date"].dt.month
)

feature_df["Quarter"] = (
    feature_df["Order_Date"].dt.quarter
)

feature_df["Day_of_Week"] = (
    feature_df["Order_Date"].dt.day_name()
)

display(feature_df)
```

### Domain-Based Features

Domain-based features rely on subject matter expertise to create variables that better reflect the business or scientific context.

These features are often highly informative because they incorporate real-world knowledge.

#### Example: Body Mass Index (BMI)

BMI combines height and weight to provide a more meaningful measure of body composition.

```python
feature_df = pd.DataFrame({
    "Weight_kg": [65, 80, 72, 90],
    "Height_m": [1.70, 1.75, 1.68, 1.82]
})

feature_df["BMI"] = (
    feature_df["Weight_kg"]
    / (feature_df["Height_m"] ** 2)
)

display(feature_df.round(2))
```

### Interpreting the Results

The examples demonstrate that meaningful features can be derived from existing variables using mathematical operations, interactions, aggregations, temporal information, and domain expertise. These engineered features often capture relationships and patterns that are not immediately apparent in the raw data, thereby enhancing analytical understanding and improving the usefulness of the dataset for subsequent analyses and predictive tasks.

### Comparing Feature Creation Techniques

| Technique | Purpose | Example |

| Mathematical Features | Create relative measures | Income per family member |
| Interaction Features | Capture joint effects | Experience × Education |
| Aggregation Features | Summarize group information | Average department salary |
| Date-Based Features | Extract temporal patterns | Month, Quarter |
| Domain-Based Features | Incorporate expert knowledge | BMI |

##Feature Selection

Feature selection is the process of identifying and retaining the most relevant variables for analysis while removing features that contribute little useful information.

The objective is to simplify datasets, reduce redundancy, improve interpretability, and enhance analytical efficiency.

Feature selection differs from feature extraction because it retains the original variables rather than transforming them into new representations.

###Why is Feature Selection Important?

Feature selection can:

- Reduce noise and irrelevant information.
- Improve model interpretability.
- Decrease computational complexity.
- Reduce overfitting.
- Improve generalization performance.
- Simplify the analytical process.

### Filter Methods

Filter methods evaluate features independently of any predictive algorithm using statistical criteria.

Features are ranked according to their relationship with the target variable, and the most informative variables are retained.

Common examples include:

- Correlation analysis
- Chi-square statistics
- Analysis of Variance (ANOVA)
- Mutual information

### Wrapper Methods

Wrapper methods evaluate subsets of features by repeatedly training and assessing predictive models.

These approaches consider interactions among variables but are often computationally expensive.

Common examples include:

- Forward Selection
- Backward Elimination
- Recursive Feature Elimination (RFE)

Features

↓

Train Model

↓

Evaluate Performance

↓

Add or Remove Features

↓

Select Best Feature Subset

### Embedded Methods

Embedded methods perform feature selection during the model training process itself.

They combine the advantages of filter and wrapper approaches by integrating selection directly into the learning algorithm.

Common examples include:

- Lasso Regression (L1 Regularization)
- Decision Trees
- Random Forest Feature Importance

### Comparing Feature Selection Methods

| Method | Selection Basis | Advantages | Limitations |
|---|---|---|---|
| Filter | Statistical measures | Fast and simple | Ignores feature interactions |
| Wrapper | Model performance | Considers interactions | Computationally expensive |
| Embedded | Training process | Balanced approach | Model dependent |

### Interpreting the Results

Feature selection helps analysts focus on the variables that contribute the greatest analytical value. While filter methods provide efficient preliminary screening, wrapper methods evaluate feature subsets more thoroughly, and embedded methods integrate selection directly into model construction.

The choice of method depends on the analytical objective, dataset size, computational resources, and the desired balance between simplicity and predictive performance.

### Important Consideration

Feature selection removes unnecessary variables while preserving the original features. In contrast, feature extraction creates entirely new representations of the data. Understanding this distinction is essential when preparing datasets for analysis and modeling.

##Introduction to Feature Extraction

Feature extraction is the process of transforming the original variables into a new set of features that capture the most important information contained in the data.

Unlike feature selection, which retains a subset of the original variables, feature extraction generates entirely new representations of the data, often with reduced dimensionality.

The primary objectives of feature extraction include:

- Reducing the number of variables.
- Eliminating redundancy.
- Improving computational efficiency.
- Preserving the most informative aspects of the data.

### Feature Selection vs Feature Extraction

| Aspect | Feature Selection | Feature Extraction |
|---|---|---|
| Purpose | Select important existing features | Create new transformed features |
| Original Variables | Retained | Transformed |
| Dimensionality Reduction | By removing variables | By generating new representations |
| Interpretability | Usually high | Often lower |
| Example | Selecting top predictors | Principal Component Analysis (PCA)|

### Principal Component Analysis (PCA)

Principal Component Analysis (PCA) is one of the most widely used feature extraction techniques.

PCA transforms correlated variables into a smaller number of uncorrelated components called principal components. These components are ordered according to the amount of variance they explain in the original data.

The first principal component captures the greatest amount of variability, followed by the second component, and so forth.

### Conceptual Illustration

Original Features

↓

Feature Standardization

↓

Principal Components

↓

Reduced Feature Space

↓

Analysis or Modeling

###Example: Applying PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pca_df = pd.DataFrame({
    "Age": [22, 25, 28, 35, 40],
    "Experience": [1, 3, 5, 10, 15],
    "Salary": [3000, 3500, 4200, 5500, 7000]
})

# Standardize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(pca_df)

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

pca_results = pd.DataFrame(
    principal_components,
    columns=["PC1", "PC2"]
)

display(pca_results)
```

```python
explained_variance = pd.DataFrame({
    "Component": ["PC1", "PC2"],
    "Explained Variance Ratio": pca.explained_variance_ratio_
})

display(explained_variance)
```

### Interpreting the Results

The principal components represent combinations of the original variables that summarize the most important information in the dataset.

The explained variance ratio indicates how much of the total variability is captured by each component. Components explaining a larger proportion of variance preserve more information from the original data.

PCA enables analysts to reduce dimensionality while retaining much of the underlying structure of the dataset.

### When Should PCA Be Used?

PCA is particularly useful when:

- The dataset contains many numerical variables.
- Features exhibit strong correlations.
- Dimensionality reduction is required.
- Computational efficiency is important.
- Visualization of high-dimensional data is desired.

### Limitations of PCA

- Principal components are often difficult to interpret.
- PCA assumes linear relationships among variables.
- Important information may be lost when too few components are retained.
- PCA is sensitive to feature scaling, making standardization an essential preprocessing step.

### Interpreting the Concept

Feature extraction focuses on representing the original data more efficiently by creating new features that summarize the underlying information. Although these transformed features may sacrifice some interpretability, they can simplify analyses and improve efficiency when working with high-dimensional datasets.
